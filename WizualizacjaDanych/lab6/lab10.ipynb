```python
fig: plt.Figure = plt.figure(figsize=(10, 7))
ax: plt.Axes = fig.gca(projection='3d')
    
count1: int = np.random.randint(10,40)
    
x: np.ndarray = np.random.randint(np.random.randint(20), np.random.randint(20,40), count1)
y: np.ndarray = np.random.randint(np.random.randint(20), np.random.randint(20,40), count1)
z: np.ndarray = np.random.randint(np.random.randint(20), np.random.randint(20,40), count1)
    
colors: np.ndarray = np.array(['r'] * 20 + ['b'] * 20 + ['g'] * 20)
markers: np.ndarray = np.array(['s'] * 25 + ['^'] * 25)

np.random.shuffle(colors)
np.random.shuffle(markers)
    
for i in range(count1):
    ax.scatter(x[i], y[i], z[i], c=colors[i], marker=markers[i])

plt.show()
```

```python
fig: plt.Figure = plt.figure(figsize=(10,7))
ax: plt.Axes = fig.gca(projection='3d')

tmp1: int = np.random.randint(10)    
tmp2: int = np.random.randint(10,20)  
      
x: np.ndarray = np.arange(tmp1, tmp2, 0.25)
y: np.ndarray = np.arange(tmp1, tmp2, 0.25)
    
x, y = np.meshgrid(x, y)
z: np.ndarray = np.sin(np.sqrt(x ** 2 + y ** 2))
    
surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm)

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
