```python
columns: List[str] = ['color','director_name','num_critic_for_reviews','duration','director_facebook_likes','actor_3_facebook_likes','actor_2_name','actor_1_facebook_likes','gross','genres','actor_1_name','movie_title','num_voted_users','cast_total_facebook_likes','actor_3_name','facenumber_in_poster','plot_keywords','movie_imdb_link','num_user_for_reviews','language','country','content_rating','budget','title_year','actor_2_facebook_likes','imdb_score','aspect_ratio','movie_facebook_likes']


movie: pd.DataFrame = pd.read_csv('movie_new.csv', sep=';', names=columns)

movie_num: pd.DataFrame = movie.select_dtypes(include=['float', 'int'])

di: Dict[str, float] = {}
    
for i in movie_num.columns:
    di[str(i)] = movie_num[i].sum() / movie_num.index.stop

movie_num: pd.DataFrame = movie_num.fillna(di)

###############################
sdv: float = 0

for i in movie_num.columns:
    if sdv < movie_num[i].std():
        sdv = movie_num[i].std()
sdv

################################

for i in movie_num.columns:
    movie[i] = movie_num[i]

################################
di2: Dict[str, str] = {}

for i in movie.columns:
    di2[str(i)] = movie[str(i)].iloc[0]

movie: pd.DataFrame = movie.fillna(di2) 

################################

col_rename: Dict[str, str]={
    'director_name': 'rezyser'
}

movie: pd.DataFrame = movie.rename(columns=col_rename)

################################

movie['imdb_score'] += 1

################################
col_rename2: Dict[str, str]={}

for i in movie.columns:
    col_rename2[i] = i.casefold().capitalize()

col_rename2

movie: pd.DataFrame = movie.rename(columns = col_rename2)

################################
a: int = 0
b: int = 0

while a >= b:
    a = np.random.randint(50)
    b = np.random.randint(50)

movie_random: pd.DataFrame = movie.iloc[a:b, :]
movie_random

################################

movie_num2: pd.DataFrame = movie.select_dtypes(include=['int','float'])

pd.options.mode.chained_assignment = None    
    
for i in movie_num2.columns:
    movie_num2[i] = movie_num2[i] + np.random.randint(100)

movie_num2
```
