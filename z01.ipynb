```python
import pandas as pd
import numpy as np
from typing import Dict

data: pd.DataFrame = pd.read_csv('dane.csv', sep='^')
    
data = data.select_dtypes(include=['number'])

def get_column_name(df: pd.DataFrame, mean: float, std: float) ->str:
     for i in df.columns:
        if round(df[str(i)].mean(), 6) == mean and round(df[str(i)].std(), 6 == std) :
            return str(i)

col_rename: Dict[str,str] ={
    get_column_name(data, 2.308642, 0.836071):'class',
    get_column_name(data, 0.383838, 2.318286):'survived',
    get_column_name(data, 446.000000, 257.353842):'passengerId',
    get_column_name(data, 29.699118, 14.526497):'age',
    get_column_name(data, 32.204208, 49.693429):'cost'}
    
data = data.rename(columns = col_rename)

data
```
