```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
movie: pd.DataFrame = pd.read_csv('movie.csv')
```

```python
groupbudget: plt.Axes = movie.groupby('title_year').agg({'budget': 'mean'}).nlargest(10,{'budget': 'mean'}).plot.bar(figsize=(10,5))
groupbudget.set_ylabel('Budget')
plt.show()
```

```python
ax: pd.DataFrame = movie.nsmallest(20,'budget')

ax.groupby('movie_title').agg({'imdb_score': 'mean'}).plot.bar(figsize=(10,5))
plt.show()
```

```python
movie['sumLikes']: pd.DataFrame = (movie['actor_1_facebook_likes'] + movie['actor_2_facebook_likes'] + movie['actor_3_facebook_likes'])
ax: plt.Axes = movie.groupby('title_year').agg({'sumLikes': 'mean'}).plot()
plt.show()
```

```python
dtf: pd.DataFrame = movie.groupby('title_year').agg({'budget': 'mean'}).query('budget.mean() < (budget + budget.std()) and budget.mean() > (budget - budget.std())')
ax: plt.Axes = dtf.plot()
plt.show()
```

```python
dtf: pd.DataFrame = movie.groupby('director_name').agg({'budget': 'mean'}).nlargest(10,'budget')
kolo: plt.Axes = dtf.plot.pie(subplots=True, autopct='%.2f %%', ylabel='', figsize=(5,10))
plt.show()
```
