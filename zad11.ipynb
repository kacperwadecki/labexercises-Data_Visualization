```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from typing import List
from typing import Dict
```

```python
domain: np.ndarray = np.arange(-10,61)
values: np.ndarray = []
    
for i in domain:
    values.append(10*i + 5)

plt.plot(domain,values, 'bs')
plt.xlabel('os x')
plt.ylabel('os y')

plt.show()
```

```python
domain: np.ndarray = np.arange(73)
values: np.ndarray = []

for i in domain:
    values.append(i * 20 - 4)

plt.plot(domain, values, 'ro-')
plt.axis([-4, 80, -1, 500])
plt.title('funkcja liniowa')
plt.xlabel('os x')
plt.ylabel('ox y')

plt.show()
```

```python
ile: int = np.random.randint(5)

for c in range(ile):
    domain: np.ndarray = np.arange(np.random.randint(50))
        
    a: int = np.random.randint(30)
    b: int = np.random.randint(30)
        
    values: np.ndarray = []
        
    for i in domain:
        values.append(i * a + b)
    
    plt.plot(domain, values, label=f'funkcja nr {c + 1}')
    plt.legend(loc='upper left')
    
plt.show()
```

```python
ile: int = np.random.randint(2,6)

for i in range(ile):
    
    domain: np.ndarray = np.arange(np.random.randint(50))
    values: np.ndarray = []
    a: int = np.random.randint(-5, 6)
    b: int = np.random.randint(15)
        
        
    for j in domain:
        values.append(j * a + b)
    
    plt.subplot(ile , 2, i + 1)
    plt.plot(domain, values, 'b')
    plt.ylabel(f'funkcja nr {i}')

plt.show()
```

```python
places: List[str] = ['Sri Lanka', 'Malediwy', 'wyspy Galapagos', 'Grecja', 'wyspy Bali', 'Hiszpania', 'Włochy']
values: List[int] = []
    
for i in range(len(places)):
    values.append(np.random.randint(1000))

plt.bar(places, values)
plt.title('siema eniu')
plt.xticks(rotation=45)

plt.show()
```
