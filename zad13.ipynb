```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from typing import List
```

```python
tips: pd.DataFrame = sns.load_dataset('diamonds')
sns.set_theme(style='darkgrid')

dataset: np.ndarray = sns.get_dataset_names()
np.random.shuffle(dataset)

randomdata: pd.DataFrame = sns.load_dataset(dataset[0]).head(20)

randomdata = randomdata.select_dtypes(include=['float','int'])

columnsdata: np.ndarray = []

for i in randomdata.columns:
    columnsdata.append(str(i))

np.random.shuffle(columnsdata)

color: str = 'blue'
    
plt.figure(figsize=(15,15))

sns.relplot(x=columnsdata[0], y=columnsdata[1], data=randomdata, marker='<', color=color)

plt.show()

print(len(randomdata))
```

```python
dataset: np.ndarray = sns.get_dataset_names()
np.random.shuffle(dataset)

randomdata: pd.DataFrame = sns.load_dataset(dataset[0]).head(5)

randomcolumns: np.ndarray = []
    
for i in randomdata.columns:
    randomcolumns.append(str(i))

np.random.shuffle(randomcolumns)

sns.countplot(x=randomcolumns[0], data=randomdata)

plt.show()
```

```python
columns: np.ndarray = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']

columns2: np.ndarray = []

number: int = np.random.randint(9)
    
for i  in range(number):
    columns2.append(columns[i])

df = pd.DataFrame(np.random.rand(number,number), columns=columns2)

sns.heatmap(df, annot=True, annot_kws={'size':10})

plt.show()
```
